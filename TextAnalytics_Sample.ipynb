```python
pos = []
for word in range(len(df_rev)):
  tag = df_rev.iloc[word, 2]
  tokens = nltk.word_tokenize(tag)
  pos_tag = nltk.pos_tag(tokens)
  pos.append(pos_tag)

df_rev['pos'] = pos
df_rev.head()
```

```python
regular_punct = list(string.punctuation)
def remove_punctuation(text,punct_list):
    for punc in punct_list:
        if punc in text:
            text = list(map(lambda x: x.replace(punc,''),text))
    return text
```

```python
pos_wo_punc = []
for word in range(len(df_rev)):
  new_label = remove_punctuation([l[0] for l in df_rev.iloc[word, 3]], regular_punct)
  pos_wo_punc.append(new_label)

df_rev['pos wo punc'] = pos_wo_punc
df_rev.head()
```

```python
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
def tokenize_pos(sent):  
        return (pos_tag(wordpunct_tokenize(sent)))
def is_stopword(token):   
        return token.lower() in set(nltk.corpus.stopwords.words('english'))
    
def is_punct(token):
    regular_punct = list(string.punctuation)
    return all([char not in list(string.punctuation) for char in token])
```

To find the most aspirational brand, I will extract positive comments and analyze them. The most mentioned brand in positive comments would be the most aspirational brand for people to buy or own.

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

I did some experiments with SentimentIntensityAnalyzer and it seems positive sentences usually have a compound polarity score of more than 0.5. The compound polarity score may exceed 0.6 for very positive sentences. Therefore, I will extract comments with a compound polarity score of more than 0.6 and look for the most mentioned brands in those comments.

```python
pos_comments = []

for i in range(len(df_rev)):
  if sia.polarity_scores(df_rev.iloc[i, 2])["compound"] > 0.6:
    pos_comments.append(df_rev.iloc[i, 2].lower())

len(pos_comments)
```

```python
pos_count = []

for elem in companies:
  num = 0
  for i in range(len(pos_comments)):
    if elem in pos_comments[i]:
      num+=1
  pos_count.append(num)

len(pos_count)
```

```python
df_aspiration = pd.DataFrame(
    {
        'company': companies,
        'frequency of mentions in positive comments': pos_count
    }
)
df_aspiration.sort_values('frequency of mentions in positive comments', ascending = False).head(10)
```

Since ac can also stand for air conditioning and man does not necessarily refer to the car manufacturer named man, the counts for these two companies (as well as some other words like king) may be unreliable. 
Therefore, it can be concluded than 'BMW' is the most aspirational brand for people to buy or own.

# Insights

From a marketing perspective, the more a brand is mentioned, the more it is mentioned with the most mentioned attributes. It can be hypothesized that more popular brands put more effort into attributes that matter to customers. In other words, identifying customer needs and focusing on them can help a brand become more popular. Also, analyzing the ratio of a brand's mentions in positive comments to its overall mentions can help segment the market and identify competitors.

From a technical point of view, distinguishing brand (like standard or man) or product (like s4) names can help improve the quality analysis and reliability of insights gained based on the results.
